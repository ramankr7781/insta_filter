```python
import cv2 as cv
cam=cv.VideoCapture(0)
while True:
    _,img=cam.read()
    img=cv.flip(img,1)
    img1=img-100
    cv.imshow("Original",img)
    cv.imshow("Filter_1(img-100)",img1)
    
    if cv.waitKey(1)==27:
        cam.release()
        break
```

```python
print("Shape of original image",img.shape)
print("Shape of filter image",img1.shape)
```

```python
print("mean of original image",img.mean())
print("mean of filter image",img1.mean())
```

```python
print("Values of original image",img)
print("Values of filter image",img1)
```

```python
import cv2 as cv
pixels=float(50)
cam=cv.VideoCapture(0)
while True:
    _,img=cam.read()
    img=cv.flip(img,1)
    img2=img*1.5
    img2[img2<0]=0
    img2[img2>255]=255
    img2=img2.astype(np.uint8)
    cv.imshow("Original",img)
    cv.imshow("Filter_2(img*1.5)",img2)
    
    if cv.waitKey(1)==27:
        cam.release()
        break
```

```python
import cv2 as cv
pixels = float(10)

cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    img = cv.flip(img, 1)
    
    img_1 = img + pixels
    img_1[img_1 <  0 ] = 0
    img_1[img_1 > 255] = 255
    img_1 = img_1.astype(np.uint8)
    
    img_2 = img*1.5
    img_2[img_2 <  0 ] = 0
    img_2[img_2 > 255] = 255
    img_2 = img_2.astype(np.uint8)
    
    img_3 = img *2
    img_3[img_3 <  0 ] = 0
    img_3[img_3 > 255] = 255
    img_3 = img_3.astype(np.uint8)
    
    
    cv.imshow("Original",img)
    cv.imshow("Filter-1",img_1)
    cv.imshow("Filter-2",img_2)
    cv.imshow("Filter-3",img_3)
    
    if cv.waitKey(1) == 27:
        cam.release()
        break
```
